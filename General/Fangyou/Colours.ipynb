```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../Data/Fangyou_data/Cleaned/Bootes_preprocessed.csv")
```

```python
fluxes = data.iloc[:,1:-11].drop(columns=["S_Code", "EBV"])
```

```python
fluxes
```

```python
number_of_columns = len(fluxes.columns)


for i in range(number_of_columns):
    for j in range(i+1, number_of_columns):
        name = fluxes.columns[i] + " - " + fluxes.columns[j]
        data[name] = fluxes.iloc[:,i] - fluxes.iloc[:,j]
```

```python
data.to_csv("../../Data/Fangyou_data/Cleaned/Bootes_preprocessed_colours.csv", index=False)
```
