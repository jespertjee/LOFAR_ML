```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../Data/Fangyou_data/Cleaned/Bootes_preprocessed.csv")
```

```python
data.columns
```

```python
fluxes = data.drop(columns=['Source_Name', "S_Code", "EBV", 'Z_BEST', 'Mass_median', 'Mass_l68',
       'Mass_u68', 'Source', 'AGN_final', 'RadioAGN_final', 'Classification',
       'Radio_excess', 'AGNfrac_af', 'AGNfrac_af_16', 'AGNfrac_cg_s_16',])
```

```python
fluxes
```

```python
def ujy_to_mag(f):
    return -2.5*np.log10(f*10**-6)
```

```python
number_of_columns = len(fluxes.columns)


for i in range(number_of_columns):
    for j in range(i+1, number_of_columns):
        name = fluxes.columns[i] + " - " + fluxes.columns[j]
        data[name] = ujy_to_mag(fluxes.iloc[:,i]) - ujy_to_mag(fluxes.iloc[:,j])
```

```python
data.to_csv("../../Data/Fangyou_data/Cleaned/Bootes_preprocessed_colours.csv", index=False)
```
