```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../../Data/Fangyou_data/Cleaned/combined_preprocessed.csv")
```

```python
fluxes = data.drop(columns=["Source_Name", "Total_flux", "Peak_flux", "S_Code", 
                            "EBV", "Z_BEST", "Mass_median", "Mass_l68", 
                            "Mass_u68", "AGN_final", "RadioAGN_final", 
                            "Classification", "Radio_excess", "AGNfrac_af", 
                            "AGNfrac_af_16", "AGNfrac_cg_s_16"])
```

```python
fluxes
```

```python
number_of_columns = len(fluxes.columns)


for i in range(number_of_columns):
    for j in range(i+1, number_of_columns):
        name = fluxes.columns[i] + " - " + fluxes.columns[j]
        data[name] = np.abs(fluxes.iloc[:,i] - fluxes.iloc[:,j])/fluxes.iloc[:,i]
```

```python
fluxes = data.drop(columns=["Source_Name", "Total_flux", "Peak_flux", "S_Code", 
                            "EBV", "Z_BEST", "Mass_median", "Mass_l68", 
                            "Mass_u68", "AGN_final", "RadioAGN_final", 
                            "Classification", "Radio_excess", "AGNfrac_af", 
                            "AGNfrac_af_16", "AGNfrac_cg_s_16"])
```

```python
means = pd.DataFrame(np.array([fluxes.mean().to_numpy(), fluxes.median().to_numpy(), fluxes.std().to_numpy()]).T, 
                     columns=["mean", "median", "std"],
                    index=fluxes.mean().index)
```

```python
means[means["median"]<0.5]["median"].dropna()
```
