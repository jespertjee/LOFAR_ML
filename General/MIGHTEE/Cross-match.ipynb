```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../../Data/MIGHTEE/Classification/final_gaussian_radio.csv')
extra_data = pd.read_csv('../../Data/MIGHTEE/Classification/Classification.txt', sep=' ')
```

```python
extra_data.columns
```

```python
data = data.drop(columns='Unnamed: 0')
data[['XAGN', 'midIRAGN', 'VLBAAGN']] = np.nan
extra_data = extra_data[['S_INT14', 'Redshift', 'RLAGN', 'XAGN', 'midIRAGN', 'optAGN', 'VLBAAGN']]
extra_data['S_INT14'] *= 10**6
```

```python
for i in data.index:
    z_original = data.loc[i, 'Z_BEST']
    flux_original = round(data.loc[i, 'S_INT14'],6)
    for k in extra_data.index:
        z_extra = extra_data.loc[k, 'Redshift']
        flux_extra = round(extra_data.loc[k, 'S_INT14'],6)
        if (z_original == z_extra) & (flux_original == flux_extra):
            print(i,k)
            data.loc[i, ['RLAGN', 'XAGN', 'midIRAGN', 'optAGN', 'VLBAAGN']] = extra_data.loc[k, ['RLAGN', 'XAGN', 'midIRAGN', 'optAGN', 'VLBAAGN']]
            extra_data = extra_data.drop(k)
            break
```
